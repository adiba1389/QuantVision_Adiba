```python
import yfinance as yf
import plotly as go
import plotly.graph_objects as py
```

```python
df=yf.download(tickers="TCS.NS",period="6mo",interval="1d")
```

```python
df.head()
```

```python
df.columns=["Close","High","Low","Open","Volume"]
df.head()
```

```python
graph=py.Figure(data=[py.Candlestick(x=df.index,open=df["Open"],high=df["High"],low=df["Low"],close=df["Close"])])
graph.show()
```

```python
graph.update_layout(xaxis_title="Date",yaxis_title="Price(INR)",title="TCS Stock Analysis using Candlestick Chart")
graph.show()
```
